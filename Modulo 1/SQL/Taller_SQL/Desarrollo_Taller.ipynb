## Parte 1: Creación de la base de datos y tabla

A continuación se muestra la creación de la base de datos `ejercicios` y una tabla manual `fuel_consumption` para alojar el dataset técnico de vehículos.

```python
CREATE DATABASE IF NOT EXISTS ejercicios;
USE ejercicios;

CREATE TABLE IF NOT EXISTS fuel_consumption (
    id INT PRIMARY KEY AUTO_INCREMENT,
    MAKE VARCHAR(50),
    MODEL VARCHAR(100),
    VEHICLECLASS VARCHAR(60),
    ENGINESIZE DECIMAL(3,1),
    CYLINDERS INT,
    TRANSMISSION VARCHAR(50),
    FUELTYPE VARCHAR(10),
    FUELCONSUMPTION_CITY DECIMAL(4,1),
    FUELCONSUMPTION_HWY DECIMAL(4,1),
    FUELCONSUMPTION_COMB DECIMAL(4,1),
    CO2EMISSIONS INT,
    MODELYEAR INT
);
```

### 7.Escribe una consulta para encontrar los vehículos que registren un consumo de combustible en ciudad inusualmente bajo o alto, aplicando filtros específicos en las columnas de consumo.

```python
SELECT MAKE, MODEL, VEHICLECLASS, FUELCONSUMPTION_CITY, CO2EMISSIONS
FROM fuel_consumption
WHERE FUELCONSUMPTION_CITY <= 6.0 
   OR FUELCONSUMPTION_CITY >= 25.0
ORDER BY FUELCONSUMPTION_CITY ASC;
```

## Parte 3: Preguntas de Análisis y Reflexión (Industria 4.0)

### 1. ¿Por qué la limpieza y filtrado de datos en SQL es el paso previo indispensable antes de entrenar un modelo de Machine Learning en Python?

La limpieza y el filtrado en SQL aseguran que los datos sean consistentes, completos y libres de valores atípicos o erróneos antes de exportarlos a Python. Si los datos contienen valores faltantes, formatos incorrectos o duplicados, el modelo de machine learning puede aprender patrones incorrectos o sesgados, lo cual reduce la calidad del entrenamiento y la precisión de las predicciones.

### 2. ¿Qué impacto tiene mapear correctamente las llaves primarias y limitar las consultas con LIMIT en lugar de traer siempre todos los datos?

Mapear correctamente las llaves primarias es fundamental para preservar la integridad referencial, evitar duplicados y permitir un acceso eficiente a filas individuales mediante índices. Limitar las consultas con `LIMIT` reduce el uso de memoria y el tiempo de respuesta en el servidor, facilita la revisión y depuración de datos, y evita cargar conjuntos de datos demasiado grandes cuando solo se necesita una muestra o los primeros registros.

### 3. ¿En qué caso específico se justificaría usar un `TRUNCATE TABLE` con la instrucción `IMMEDIATE` en lugar de un `DELETE` tradicional?

Un caso justificable para usar `TRUNCATE TABLE ... IMMEDIATE` es cuando se necesita limpiar rápidamente una tabla temporal o de staging en un pipeline de datos en tiempo real, y no es necesario mantener un historial de transacciones ni activar triggers de eliminación. Esta operación es útil cuando se desea reiniciar el contenido con la misma estructura de tabla, con el menor costo de registro posible, garantizando que el espacio sea liberado de forma inmediata.
