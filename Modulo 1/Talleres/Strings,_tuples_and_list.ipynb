¿Cuál es el valor de la variable a después de ejecutar el siguiente código?

```python
# La variable a recibe el texto "1".
# El valor de a es una cadena de texto (str).
a = "1"
print(a)
print(type(a))  # <class 'str'>
```

Considere la siguiente tupla:

```python
# sample tuple

genres_tuple = ("pop", "rock", "soul", "hard rock", "soft rock", \
                "R&B", "progressive rock", "disco")
genres_tuple
```

Acceda al elemento, con respecto al índice 3:

```python
# Accedemos al elemento en el índice 3 de la tupla
# Los índices comienzan en 0, entonces el elemento 3 es el cuarto.
print(genres_tuple[3])
```

Utilice el segmentado para obtener los índices 3, 4 y 5:

```python
# Utilizamos slicing para obtener los índices 3, 4 y 5
print(genres_tuple[3:6])
```

Encuentra los dos primeros elementos de la tupla genres_tuple:

```python
# Tomamos los dos primeros elementos de la tupla
print(genres_tuple[:2])
```

Encuentra el primer índice de "disco":

```python
# Encontramos el índice de la cadena 'disco' dentro de la tupla
print(genres_tuple.index('disco'))
```

Generar una lista ordenada a partir de la tupla C_tuple=(-5, 1, -3):

```python
# Convertimos la tupla en una lista ordenada
C_tuple = (-5, 1, -3)
C_list = sorted(C_tuple)
print(C_list)
```
