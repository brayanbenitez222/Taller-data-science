¿Cuál es el tipo de dato del resultado de: 6 / 2?

```python
# División normal en Python retorna un número de punto flotante
resultado = 6 / 2
print(resultado)
print(type(resultado))  # <class 'float'
```

¿Qué tipo de resultado tiene la operación: 6 // 2?

```python
# División entera en Python retorna un entero cuando los operandos son divisibles
resultado = 6 // 2
print(resultado)
print(type(resultado))  # <class 'int'>
```

¿Cuál es el tipo de resultado de: "Hola, mundo!"?

```python
# En Python, un texto entre comillas es una cadena de caracteres (str)
texto = "Hola, mundo!"
print(texto)
print(type(texto))  # <class 'str'>
```

Escribe el código para convertir el siguiente número que representa el ID de empleado "1001" a un número entero.

```python
# Convertimos una cadena que representa un número entero a un valor int
documento_empleado = "1001"
numero_empleado = int(documento_empleado)
print(numero_empleado)
print(type(numero_empleado))  # <class 'int'>
```

Escribe el código para convertir este número que representa el valor financiero "1234.56" a un número de punto flotante.

```python
# Convertimos una cadena que representa un número decimal a float
valor_financiero = "1234.56"
importe = float(valor_financiero)
print(importe)
print(type(importe))  # <class 'float'>
```

Escribe el código para convertir este número de teléfono 123-456-7890 a una cadena de texto.

```python
# Convertimos un número de teléfono a cadena de texto
telefono = "123-456-7890"
telefono_str = str(telefono)
print(telefono_str)
print(type(telefono_str))  # <class 'str'>
```

Escribe una expresión para sumar 30 y 20 y restar 40.

```python
# Sumamos 30 y 20, y luego restamos 40
resultado = 30 + 20 - 40
print(resultado)
print(type(resultado))  # <class 'int'>
```

Escribe una expresión para restar 5 de 55 y dividir el resultado entre 10.

```python
# Restamos 5 de 55 y dividimos el resultado entre 10
resultado = (55 - 5) / 10
print(resultado)
print(type(resultado))  # <class 'float'>
```

Escribe una expresión para multiplicar 6 por 10 y dividir el resultado entre 12.

```python
# Multiplicamos 6 por 10 y dividimos entre 12
resultado = (6 * 10) / 12
print(resultado)
print(type(resultado))  # <class 'float'>
```

¿Cuál es el valor de x donde x = 3 + 2 * 2?

```python
# x se calcula con la prioridad de multiplicación primero
x = 3 + 2 * 2
print(x)
print(type(x))  # <class 'int'>
```

¿Cuál es el valor de y donde y = (3 + 2) * 2?

```python
# y se calcula con paréntesis que cambian el orden de operaciones
y = (3 + 2) * 2
print(y)
print(type(y))  # <class 'int'>
```

¿Cuál es el valor de z donde z = x + y?

```python
# Sumamos x y y para obtener z
x = 3 + 2 * 2
y = (3 + 2) * 2
z = x + y
print(z)
print(type(z))  # <class 'int'>
```
