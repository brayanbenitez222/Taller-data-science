## Cuestionario sobre Diccionarios


<b>Necesitarás este diccionario para las siguientes dos preguntas:</b>


````python
# Diccionario de muestra para las preguntas

soundtrack_dic = {"The Bodyguard":"1992", "Saturday Night Fever":"1977"}
soundtrack_dic
````

a) En el diccionario <code>soundtrack_dic</code>, ¿cuáles son las llaves (keys)?


````python
# Escribe tu código abajo y presiona Shift+Enter para ejecutar
soundtrack_dic.keys()
# El método .keys() retorna todas las llaves del diccionario. En este caso: "The Bodyguard" y "Saturday Night Fever". Las llaves son los identificadores únicos de cada entrada.
````

<details><summary>Haz clic aquí para ver la solución</summary>

```python
soundtrack_dic.keys() # Las llaves son "The Bodyguard" y "Saturday Night Fever"

```

</details>


b) En el diccionario <code>soundtrack_dic</code>, ¿cuáles son los valores (values)?


````python
# Escribe tu código abajo y presiona Shift+Enter para ejecutar
soundtrack_dic.values()
# El método .values() retorna todos los valores del diccionario. Son los años "1992" y "1977" — nótese que son strings, no enteros, porque están entre comillas.
````

<details><summary>Haz clic aquí para ver la solución</summary>

```python
soundtrack_dic.values() # Los valores son "1992" y "1977"

```

</details>

## Escenario: Tienda de Inventario
    


El proyecto del escenario de la tienda de inventario utiliza un enfoque basado en diccionarios para desarrollar un sistema sólido para gestionar y realizar el seguimiento del inventario en una tienda minorista.

**Nota: Trabajarás con los detalles de dos productos.**


## Tarea-1 Crear un diccionario vacío



Primero necesitas crear un diccionario vacío, donde almacenarás los detalles de los productos.


````python
# Escribe tu código aquí
inventory = {}
# Un diccionario vacío se crea con llaves vacías {}. Luego se puede ir llenando con pares llave-valor usando diccionario["llave"] = valor.
````

<details> <summary>Haz clic aquí para ver la solución</summary>

```python

inventory ={}

```

</details>


## Tarea-2 Almacenar los detalles del primer producto en variables
* Nombre del Producto (Product Name) = Mobile phone
* Cantidad del Producto (Product Quantity) = 5
* Precio del Producto (Product price) = 20000
* Año de Lanzamiento del Producto (Product Release Year) = 2020


````python
# Escribe tu código aquí
ProductNo1 = "Mobile Phone"
ProductNo1_quantity = 5
ProductNo1_price = 20000
ProductNo1_releaseYear = 2020
# Guardamos cada dato del producto en una variable separada. Esto facilita agregar los datos al diccionario en el siguiente paso. El nombre es string, los demás son enteros.
````

<details> <summary>Haz clic aquí para ver la solución</summary>

```python
ProductNo1 = "Mobile Phone"
ProductNo1_quantity = 5
ProductNo1_price = 20000
ProductNo1_releaseYear= 2020
```

</details>


## Tarea-3 Agregar los detalles al inventario


````python
# Escribe tu código aquí
inventory["ProductNo1"] = ProductNo1
inventory["ProductNo1_quantity"] = ProductNo1_quantity
inventory["ProductNo1_price"] = ProductNo1_price
inventory["ProductNo1_releaseYear"] = ProductNo1_releaseYear
# Para agregar entradas a un diccionario existente usamos diccionario["nueva_llave"] = valor. Cada línea agrega un atributo del producto como una entrada separada.
````

<details> <summary>Haz clic aquí para ver la solución</summary>

```python

inventory["ProductNo1"]= ProductNo1
inventory["ProductNo1_quantity"]= ProductNo1_quantity
inventory["ProductNo1_price"]= ProductNo1_price
inventory["ProductNo1_releaseYear"]=ProductNo1_releaseYear

```

</details>
